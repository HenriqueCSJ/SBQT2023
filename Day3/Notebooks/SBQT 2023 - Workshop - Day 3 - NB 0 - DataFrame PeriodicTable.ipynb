```python
import pandas as pd
from mendeleev import element
```

```python
# List to store the data
data_list = []
```

```python
# Loop over all 118 elements and gather data
for i in range(1, 119):
    el = element(i)
    data = {
        "Atomic Number": el.atomic_number,
        "Name": el.name,
        "Symbol": el.symbol,
        "Atomic Mass (u)": el.atomic_weight,
        "Atomic Radius (pm)": el.atomic_radius,
        "Covalent Radius Pyykko (pm)": el.covalent_radius_pyykko if el.covalent_radius_pyykko else "N/A",
        "Electronegativity (Ghosh)": el.en_ghosh if el.en_ghosh else "N/A",
        "Electronegativity (Pauling)": el.en_pauling if el.en_pauling else "N/A"
    }
    data_list.append(data)
```

```python
# Create a DataFrame
df = pd.DataFrame(data_list)
```
