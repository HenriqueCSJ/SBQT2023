# Plot data for low dimensional molecules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
import plotly.graph_objects as go
```

```python
df = pd.read_csv("../Databases/2D_molecule.csv")
```

```python
df = df[['Structure name',
 'Bond lengths 1',
 'Bond lengths 2',
 'Bond angles',
 'CAS Heff D']]
```

```python
df = df.dropna()
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(x="Bond angles", y="CAS Heff D",data=df, hue="Structure name")
plt.savefig("2Dmolecules.png", dpi=600)
```

```python
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
```

```python
x = df["Bond angles"].values
y = df["Bond lengths 1"]
z = df["CAS Heff D"]

fig = go.Figure(data=[go.Mesh3d(x=x, y=y, z=z, opacity=1, intensity=z, colorscale="Viridis")])
fig.show()
plt.savefig("2Dmolecules.png", dpi=600)
```
