# Cosine Similarity in Word Embeddings

The scores are calculated using **cosine similarity**, a metric used to determine how similar two vectors are. Specifically, for word embeddings, it measures the cosine of the angle between two vectors in a multi-dimensional space. The formula for cosine similarity between two vectors \( A \) and \( B \) is:

$$
\text{cosine similarity}(A, B) = \frac{A \cdot B}{\|A\| \times \|B\|}
$$

Where:
- $( A \cdot B )$ is the dot product of the vectors.
- $( \|A\| )$ and $( \|B\| )$ are the magnitudes (or lengths) of vectors \( A \) and \( B \), respectively.

The result will be a value between -1 and 1:
- A value of **1** means the vectors are identical in orientation (they point in the same direction).
- A value of **0** means the vectors are orthogonal (they are not pointing in the same or opposite direction, but are perpendicular).
- A value of **-1** means the vectors are diametrically opposed (they point in exactly opposite directions).

For word embeddings, the vectors represent words, and the cosine similarity score indicates how similar the context or meaning of the words is based on their occurrences in the training data. In general:
- A high positive score (close to 1) indicates the words are used in similar contexts.
- A score close to 0 indicates the words are not particularly related.
- A negative score (which is rare in typical word2vec models) would indicate opposing contexts.

In the case of the **word2vec** model, the vectors for each word are generated such that words appearing in similar contexts have vectors pointing in similar directions. Thus, by calculating the cosine similarity between two word vectors, we can gauge how semantically or contextually similar those words are.
